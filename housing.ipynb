```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

df = pd.read_csv('housing.csv')
df.head()
```

```python

# Create a histogram of median house value with 30 bins
plt.figure(figsize=(10, 6))
sns.histplot(df['median_house_value'], bins=30, kde=False, color='lightblue')

# Set labels and title
plt.title('Histogram of Median House Values', fontsize=16)
plt.xlabel('Median House Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Set the x-axis limits to match the range in the example image
plt.xlim(200000, 500000)

# Show the plot
plt.show()

```
